```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import animation
import glob
%matplotlib inline
```

```python
im_h = 20
im_w = 20
tm_l = 250
target = './20x20_point/'
```

```python
phie = np.zeros((tm_l , im_h, im_w), dtype=np.float32)
vmem = np.zeros((tm_l , im_h, im_w), dtype=np.float32)
for i, path in enumerate(glob.glob(target+'vmem_*.npy')[:tm_l]) : vmem[i, :, :] = np.load(path)
for i, path in enumerate(glob.glob(target+'phie_*.npy')[:tm_l]) : phie[i, :, :] = np.load(path)
```

```python
ts_phie = phie[:, im_h/2, im_w/2]
ts_vmem = vmem[:, im_h/2, im_w/2]
plt.plot(ts_vmem)
#plt.plot(ts_phie)
```

```python
fig = plt.figure(figsize=(10,10))
list_plot = range(0, 100, 10)
print list_plot
for i, t in enumerate(list_plot):
    plt.subplot( len(list_plot), 2, i * 2 + 1)
    plt.imshow( vmem[t,:,:], cmap='gray', interpolation='nearest', vmin=-100, vmax=100)
    plt.subplot( len(list_plot), 2, i * 2 + 2)
    plt.imshow( phie[t,:,:], cmap='gray', interpolation='nearest', vmin=-200, vmax=0)
plt.show()
```
